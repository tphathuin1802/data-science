```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Tải dữ liệu bản đồ của Hoa Kỳ
usa = gpd.read_file(gpd.datasets.get_path('https://www.naturalearthdata.com/downloads/110m-cultural-vectors/')).query("continent == 'North America'")

# Lọc chỉ lấy nước Mỹ
usa = usa[usa['name'] == 'United States']

# Đọc shapefile của các bang của Mỹ
states = gpd.read_file("https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json")

# Chọn tiểu bang California (hoặc thay đổi theo yêu cầu)
selected_state = states[states['name'] == 'California']

# Vẽ bản đồ
fig, ax = plt.subplots(figsize=(10, 6))
states.plot(ax=ax, color='lightgray', edgecolor='black')  # Bản đồ các bang
selected_state.plot(ax=ax, color='red')  # Bang được chọn tô màu đỏ

# Ẩn trục
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Hiển thị bản đồ
plt.show()
```
