```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
print("finally import the data!")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from tensorflow.keras.layers import Conv2D


img = train_images[0]
print(f"Default image shape: {img.shape}")

#Standardize the data
img = img / 255.0

#change to array for easier manipulation:

img = np.array(img)

img_reshaped = img.reshape((1, 28, 28, 1))

print(f"Image shape after conversion: {img_reshaped.shape}")
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
conv_layer = tf.keras.layers.Conv2D(
    filters =8,
    kernel_size=(3, 3)
)

conv_layer.build(img_reshaped.shape)
```

```python
output_image = conv_layer(img_reshaped)

output_image_np = output_image.numpy().squeeze()
print(f"Output image shape: {output_image_np.shape}")
```

```python
fig, axs = plt.subplots(3,3, figsize=(10, 10))
axs = axs.flatten()

axs[0].imshow(img, cmap='gray')
axs[0].axis('off')
axs[0].set_title("Original Image")

for i in range(8):
    axs[i+1].imshow(output_image_np[:, :, i], cmap='gray')
    axs[i+1].axis('off')
    axs[i+1].set_title(f"Filter {i+1}")
```
